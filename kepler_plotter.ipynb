```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import matplotlib
matplotlib.rcParams.update({'font.size':15})
```

# Orbit of Mercury

```python
x,y,v_x,v_y,r,t= np.loadtxt("mercury_positions.dat",unpack=True)
xp,yp,rp,thetap,tp = np.loadtxt("values_precession.dat",unpack=True)
```

```python
zeros = np.zeros(len(xp))
plt.figure(figsize=(8,8))
plt.plot(x,y)
plt.xlabel("x [AU]")
plt.ylabel("y [AU]")
plt.xlim(-0.5,0.5)
plt.ylim(-0.5,0.5)
plt.axvline(0,linestyle="dashed",color="k")
plt.axhline(0,linestyle="dashed",color="k")
for i in range(len(xp)):
    plt.plot([zeros[i],xp[i]],[zeros[i],yp[i]],"-",markersize=10,label="t={0} yr".format(round(tp[i],2)))
#plt.axvline(0.47)
#plt.axhline(0.47)
plt.title(r"Mercury's perihelion precession: $\alpha = 0.005$")
plt.legend(loc=3)
#plt.tight_layout()
plt.savefig("alpha_0-05_mercury_prec.png")
plt.show()
```

# Radius of Mercury as a function of time

```python
matplotlib.rcParams.update({'font.size':20})
plt.figure(figsize=(15,5))
plt.plot(t,r)
plt.axhline(r[0],linestyle="dashed")
plt.xlabel("time [yr]")
plt.ylabel("radius [AU]")
#plt.xlim(6,9)
plt.plot(tp,rp,".",markersize=30)
plt.tight_layout()
plt.savefig("radius_time.png")
plt.show()
```

# Variation of the radius of Mercury dr

```python
dr = []
dt = []
for i in range(len(r)-1): 
    dr.append((r[i+1]-r[i])*100+0.3)
    dt.append(t[i+1]-t[i])
```

```python
plt.figure(figsize=(15,5))
plt.plot(t,r,label="r")
plt.axhline(r[0],linestyle="dashed")
#plt.xlabel("time [yr]")
#plt.ylabel("radius [AU]")
#plt.show()
plt.plot(t[0:-1],dr,label="dr*100+0.3")
plt.axhline(0.3)
plt.legend()
plt.show()
```

zeros of dr are maximum and minimum of r

# Precession angle as a function of time

```python
plt.plot(tp,10.97*tp-1.18)
plt.plot(tp,thetap,".",markersize=15)
plt.xlabel("time [yr]")
plt.ylabel(r"$\theta$ [deg]")
plt.tight_layout()
plt.savefig("theta_time.png")
plt.show()
```

Be careful that if the angle of precession becomes greater than 90 degrees, the tangent becomes negative and the angles become negative too
